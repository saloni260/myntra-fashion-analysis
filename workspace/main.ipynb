```python
import matplotlib.pyplot as plt

# Define price bins
price_bins = [0, 500, 1000, 1500, 2000, 5000, 10000, float('inf')]
labels = ['0-500', '500-1000', '1000-1500', '1500-2000', '2000-5000', '5000-10000', '10000+']

# Small brands
small_brand_dress_df = dress_df[dress_df['BrandName'].isin(small_brand_names)]
small_counts = pd.cut(small_brand_dress_df['DiscountPrice (in Rs)'], bins=price_bins, labels=labels).value_counts().sort_index()

# Big brands
big_brand_names = brand_counts[brand_counts > 5].index.tolist()
big_brand_dress_df = dress_df[dress_df['BrandName'].isin(big_brand_names)]
big_counts = pd.cut(big_brand_dress_df['DiscountPrice (in Rs)'], bins=price_bins, labels=labels).value_counts().sort_index()

# Plot pie charts
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

axes[0].pie(small_counts, labels=labels, autopct='%1.1f%%', startangle=140)
axes[0].set_title('Price Range Distribution - Small Brands (Dresses)')

axes[1].pie(big_counts, labels=labels, autopct='%1.1f%%', startangle=140)
axes[1].set_title('Price Range Distribution - Big Brands (Dresses)')

plt.tight_layout()
plt.show()
```
